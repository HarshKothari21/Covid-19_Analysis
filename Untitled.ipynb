```python
counties = df['location'].unique()
```

```python
filt = df['location'].isin(countries)
mydf = df[filt]
```

```python
mydf = mydf.sort_values(by=['total_cases'], ascending=[False])
value = mydf.iloc[0, 4]
value
```

```python
max_cases = []
for i in countries:
    temp_flt = mydf['location'] == i
    temp=mydf.loc[temp_flt]
    max_cases.append(temp.iloc[0, 4])
max_cases  
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.barh(countries, max_cases)

plt.gcf().autofmt_xdate()
plt.title("Country Analysis")
plt.xlabel("Number of Positive cases")
plt.tight_layout()
plt.show()
```
